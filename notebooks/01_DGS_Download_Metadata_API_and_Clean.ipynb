```python
def download_DGS_metadata_(bid):
    url = API + str(bid)
    print "Downloading %3d - %s" % (bid, url)
    r = requests.get(url)
    if r.status_code == requests.codes.ok:
        #return r.text.encode('ascii',errors='ignore')
        return r.json()
    else:
        return None
```

```python
def download_DGS_metadata(bid):
    isDone   = False
    res      = None
    attempts = 1    
    while(not isDone and attempts <= 5):
        try:
            res = download_DGS_metadata_(bid)
            isDone = True
        except Exception as e:
            attempts += 1
            print '\tException: %s, retrying - attempt %d' % (e, attempts)
    return res
```

```python
def load_DGS_metadata(raw_dir, filename):
    try:
        #print "Loading " + filename
        filename = os.path.join(raw_dir, filename)        
        with open(filename) as data_file:
            return data_file.read()
    except Exception as e:
        print "Loading metadata for building# " + str(bid) + " Failed! " + str(e)
        return None
```

```python
# Load and format the metadata
metadata = []
filenames = os.listdir(raw_dir)
for filename in filenames:    
    if filename.endswith('.json'):
        print "Loading " + filename, 
        dat = load_DGS_metadata(raw_dir, filename)
        if dat == None:
            print '\t Skipping...'
            pass
        try:
            #dat1 = dat.encode('ascii',errors='ignore').decode('ascii')
            dat1 = ''.join([i if ord(i) < 128 else ' ' for i in dat])    ## unicode to ascii        
            #print isinstance(dat, unicode), isinstance(dat1, unicode), 
            jsDict = json.loads(dat1)
            jsDict = flatdict.FlatDict(jsDict)
            print '\t %6d attributes' % (len(jsDict.keys()))            
            metadata.append(jsDict)
        except Exception as e:
            print "Loading metadata failed! " + filename + " " + str(e)
```

```python
df3 = df2.copy()

#remove the $sign in annual_energy_cost
df3.loc[:, 'annual_energy_cost'] = df3.loc[:, 'annual_energy_cost'].str.replace('$', '')

# If square_feet in NaN, then use pm_property_information:grossFloorArea
df3['square_feet'] = df3.apply(
    lambda row: row['grossFloorArea'] if np.isnan(row['square_feet']) and pd.to_numeric(row['grossFloorArea']) > 2 else row['square_feet'],
    axis=1
)

# If square_feet in NaN, then use pm_property_information:grossFloorArea
df3['PSU'] = df3.apply(
    lambda row: row['primaryFunction'] if not pd.isnull(row['primaryFunction']) and row['PSU'] == 'Other - Public Services' else row['PSU'],
    axis=1
)

# add industry column
industries = [ industry[bt] if bt in industry.keys() else 'Unknown' for bt in df3['PSU'] ]
df3.insert(2, 'Industry', industries)

# save this
filename = os.path.join(processed_dir, 'dgs_metadata_filtered_and_cleaned.csv')        
df3.to_csv(filename, float_format='%.02f', encoding="utf-8")
print df3.shape, filename

#df3.head()
```
