```python
features = []

for filename in feature_file_list:
    try:
        df = pd.read_csv(os.path.join(repos_path, "data/processed/", filename), index_col="building_name")
        print filename
        features.append(df)
    except:
        continue
```

```python
features = pd.concat(features, axis=1)
```

```python
features.T.head(100)
```

```python
features.columns = features.columns.str.replace('[^\w\s]','')
```

```python
features = features[features.columns.sort_values()]
```

```python
features = features[features.columns.sort_values()]
```

```python
features.to_csv(os.path.join(repos_path, "data/processed/ALLFEATURES.CSV"))
```
